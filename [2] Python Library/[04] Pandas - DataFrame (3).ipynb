apply()ëŠ” í–‰ê³¼ ì—´ì— ë²¡í„°í™” ì—°ì‚°ì„ í•  ìˆ˜ ìˆë„ë¡ í•´ì£¼ëŠ” í•¨ìˆ˜ì´ë‹¤.<br>
ì¼ë°˜ì ìœ¼ë¡œ ë³µì¡í•œ ì—°ì‚°ì„ í•´ê²°í•˜ê¸° ìœ„í•œ lambda í•¨ìˆ˜ë‚˜ ì‚¬ìš©ì ì •ì˜ í•¨ìˆ˜ë¥¼ ê° Row ë˜ëŠ” Columnì— ì¼ê´„ ì ìš©ì‹œí‚¤ê¸° ìœ„í•´ ì‚¬ìš©í•œë‹¤.<br>

```python
df3 = pd.DataFrame({
    'a':[1, 3, 4, 3, 4],
    'b':[2, 3, 1, 4, 5],
    'c':[1, 5, 2, 4, 4]
})

df3
```

```python
df3[['a', 'b']].apply(np.square)
```

ì‚¬ìš©ì ì •ì˜ í•¨ìˆ˜ë¥¼ apply() í•¨ìˆ˜ì— ì‚¬ìš©í•  ìˆ˜ ìˆë‹¤.

```python
df3.apply(diff, axis=0)
```
