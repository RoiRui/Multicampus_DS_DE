## ğŸ“‚ DataFrame Concate

ConcateëŠ” ë‘ ê°œì˜ DataFrameì„ ê¸°ì¤€ ì—´ ì—†ì´ ë³‘í•©í•œë‹¤.<br>
ë‹¨ìˆœíˆ ë‘ Series ë˜ëŠ” DataFrameì„ ë³‘í•©í•˜ê¸° ë•Œë¬¸ì— ì¸ë±ìŠ¤ê°€ ì¤‘ë³µë  ìˆ˜ë„ ìˆë‹¤.

```python
df1 = pd.DataFrame({'A':['A0', 'A1', 'A2', 'A3'], 
                    'B':['B0', 'B1', 'B2', 'B3'],
                    'C':['C0', 'C1', 'C2', 'C3'],
                    'D':['D0', 'D1', 'D2', 'D3']
                   }, index=[0, 1, 2, 3])
df1
```

```python
df2 = pd.DataFrame({'A':['A4', 'A5', 'A6', 'A7'],
                    'B':['B4', 'B5', 'B6', 'B7'],
                    'E':['C4', 'C5', 'C6', 'C7'],
                    'F':['D4', 'D5', 'D6', 'D7']
                   }, index=[0, 1, 2, 3])
df2
```

```python
df3 = pd.DataFrame({'A':['A8', 'A9', 'A10', 'A11'],
                    'B':['B8', 'B9', 'B10', 'B11'],
                    'C':['C8', 'C9', 'C10', 'C11'],
                    'G':['D8', 'D9', 'D10', 'D11']
                   }, index=[1, 2, 3, 4])
df3
```

```python
pd.concat([df1, df2, df3])
```
