## üìÇ Line Plot ÏòàÏ†ú

```python
x = [0, 1, 2, 3, 4, 5, 6]
y = [1, 4, 5, 8, 9, 5, 3]

plt.plot(x, y, color='green', linestyle='dashed', marker='o', markerfacecolor='blue', markersize=12)
plt.show()
```

```python
plt.plot([10, 20, 30, 40], [1, 4, 9, 16], c='b', lw=5, ls='--', marker='o', ms=15, mec='g',mew=5, mfc='r')
plt.xlim(0, 50)
plt.ylim(-10, 30)
plt.title('ÌÉÄÏù¥ÌãÄ')
```

```python
x = np.arange(0., 5., 0.2) # ÌïòÎÇòÏùò Ï∞®Ìä∏Ïóê Ïó¨Îü¨ ÏÑ†ÏùÑ ÌëúÌòÑ
plt.plot(x, x, 'r--') # red line, dashed line
plt.plot(x, 0.5 * x ** 2, 'gs:') #blue line, dot line, square marker
plt.plot(x, 0.2 * x ** 3, 'b^-') #green line, solid line, triangle_up marker
plt.show()
```

```python
plt.plot(x, x, 'r--', x, 0.5 * x ** 2, 'gs:', x, 0.2 * x ** 3, 'b^-') # ÏúÑ Í∑∏ÎûòÌîÑÏùò ÏΩîÎìúÎ•º plot() Ìïú Î≤àÏóê ÌëúÌòÑ
plt.show()
```

```python
x = [10, 20, 30, 40, 50, 60]
y = [11, 15, 40, 40, 20, 10]
plt.title('ÌÉÄÏù¥ÌãÄ', loc='right', pad=15, fontsize=20, fontweight='bold')
plt.plot(x, y, color='green', linestyle='dashed', marker='o')
plt.xticks(x, labels=['10ÎåÄ','20ÎåÄ','30ÎåÄ','40ÎåÄ','50ÎåÄ','60ÎåÄ'])
plt.yticks([0, 10, 20, 30, 40, 50])
plt.grid(True)
plt.show()
```

```python
x = np.arange(0, 12, 0.1)
y = np.sin(x)

plt.plot(x, y, c='blue')
plt.plot(x, np.cos(x), c='red')
plt.xlabel('time')
plt.ylabel('Amplitude')
plt.grid()
plt.show()
```

## üìÇ Box Plot ÏòàÏ†ú

```python
s1 = np.random.normal(loc=0, scale=1, size=1000)
s2 = np.random.normal(loc=10, scale=0.5, size=1000)
s3 = np.random.normal(loc=20, scale=2, size=1000)

plt.plot(s1, label='s1')
plt.plot(s2, label='s2')
plt.plot(s3, label='s3')
plt.legend()
plt.show()
```

```python
plt.boxplot([s1, s2, s3])
plt.grid()
plt.show()
```
